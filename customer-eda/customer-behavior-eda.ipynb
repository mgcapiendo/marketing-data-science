# Load Data

```python
from ucimlrepo import fetch_ucirepo

# fetch dataset
online_retail = fetch_ucirepo(id=352)

# data (as pandas dataframes)
X = online_retail.data.features
y = online_retail.data.targets

# metadata
print(online_retail.metadata)

# variable information
print(online_retail.variables)
```

```python
df = X.copy()
print(df.shape)
df.head()
```

```python
# we need the two index columns invoiceno and stockcode, replacing df to be the original complete dataset
df = online_retail.data.original.copy()
print(df.shape)
df.head()
```

```python
df.info()
```

# Clean Data

```python
# drop the rows with no customer ID since we cant track them
df = df.dropna(subset=['CustomerID'])

# change invoicedata from object to datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# check if rows got dropped from 541909
print("Remaining Rows:", df.shape[0])

print("InvoiceDte type is no longer an object and is now:", df['InvoiceDate'].dtype)
```

# Feature Engineering - Total Spent

```python
# make a total spent column 
df['TotalSpent'] = df['Quantity'] * df['UnitPrice']
```

# Analysis - Top Customer

```python
# find top customers by sorting through grand total per customer across all purchases
#group first by individual customer, then find the totalspend column of that customer, sum their total spent for a grand total, sort from biggest to smallest, show the top 10
top_customers = df.groupby('CustomerID')['TotalSpent'].sum().sort_values(ascending=False).head(10)
print(top_customers)
```

# Visualization - Top Customer

```python
# visual of top 10 customers by total spent
plt.figure(figsize=(12,6))
# set number labels to string for safety
sns.barplot(x=top_customers.index.astype(str), y=top_customers.values)

plt.title('Top 10 Customers by Total Spent')

plt.xlabel('CustomerID')
plt.ylabel('Total Spent in Euros')

plt.xticks(rotation=0)

plt.tight_layout()
plt.show()
```

# Analysis - Top Country

```python
# number of countries
print('Number of Countries:', df['Country'].nunique())
print(df['Country'].unique())
```

```python
# find out what the top country is in total spent
top_country = df.groupby('Country')['TotalSpent'].sum().sort_values(ascending=False).head(5)
print(top_country)
```

# Visualization - Top Country

```python
# visual of top 5 countries by total spent
plt.figure(figsize=(12,6))

sns.barplot(x=top_country.index.astype(str), y=top_country.values)

plt.title('Top 5 Countries by Total Spent')

plt.xlabel('Country')
plt.ylabel('Total Spent in Euros')

plt.xticks(rotation=0)

plt.ticklabel_format(style='plain', axis='y')

plt.tight_layout()
plt.show()
```

# Analysis - Most Cancelled Purchase

```python
# how many purchases get cancelled
cancelled = df[df['InvoiceNo'].str.startswith('C')]

print('Number of cancellations:', len(cancelled))
```

```python
# find out what the top cancelled purchase is
# cancelled are negative meaning -1 is the "biggest", need to sort from smallest to largest for most cancelled purchase
top_cancelled = cancelled.groupby('Description')['Quantity'].sum().sort_values(ascending=True).head(5)
print(top_cancelled)
```

# Visualization - Top Cancelled Purchase

```python
# fix negative numbers 
top_cancelled_fix = top_cancelled.abs()
```

```python
# visual of top 5 cancelled purchases
plt.figure(figsize=(12,6))

sns.barplot(x=top_cancelled_fix.index.astype(str), y=top_cancelled_fix.values)

plt.title('Top 5 Cancelled Purchases')

plt.xlabel('Item')
plt.ylabel('Cancellations')

plt.xticks(rotation=90)

# plt.ticklabel_format(style='plain', axis='y')

plt.tight_layout()
plt.show()
```

# Feature Engineering - Month

```python
# extract just the month from the date and time
df['Month'] = df['InvoiceDate'].dt.month
```

# Analysis - Top Month Profit

```python
# figure out top month in total spent, sort by index for calendar
top_month = df.groupby('Month')['TotalSpent'].sum().sort_index(ascending=True).head(12)
print(top_month)
```

# Visualization - Top Month Profit

```python
# visual of profit by months
plt.figure(figsize=(12,6))

sns.barplot(x=top_month.index.astype(str), y=top_month.values)

plt.title('Purchaes by Month')

plt.xlabel('Month Number')
plt.ylabel('Purchases')

plt.xticks(rotation=0)

plt.ticklabel_format(style='plain', axis='y')

plt.tight_layout()
plt.show()
```
